```python
# Ubah kolom 'hour' menjadi format waktu

df['hour'] = df['hour'].astype(str).apply(lambda x: x.zfill(2) + '.00' if '.' not in x else x)

# Now convert to datetime objects, including seconds in the format
df['hour'] = pd.to_datetime(df['hour'], format='%H.%M').dt.time
# Format string updated to include seconds and fractional seconds


# Define a function to


# Define a function to classify hours into morning, afternoon, or evening periods
def classify_period(hour):
    if hour >= pd.to_datetime('06:30').time() and hour <= pd.to_datetime('08:30').time():
        return 'Pagi'
    elif hour >= pd.to_datetime('11:30').time() and hour <= pd.to_datetime('13:30').time():
        return 'Siang'
    elif hour >= pd.to_datetime('14:00').time() and hour <= pd.to_datetime('18:00').time():
        return 'Sore'

df['periode'] = df['hour'].apply(classify_period)

# Group by longitude_latitude and periode, then aggregate traffic values into a list
pivot_df = df.groupby(['longitude_latitude', 'periode'])['traffic'].apply(list).reset_index()

# Convert the list of traffic data into individual columns (traffic1, traffic2, ...)
traffic_df = pivot_df['traffic'].apply(pd.Series)
traffic_df.columns = [f'traffic{i+1}' for i in range(traffic_df.shape[1])]

# Concatenate the longitude_latitude, periode, and the traffic data
final_pivot = pd.concat([pivot_df[['longitude_latitude', 'periode']], traffic_df], axis=1)

# Save the final pivot table to Excel
output_path = '/content/output_pivot_table.xlsx'
final_pivot.to_excel(output_path, index=False)

print("Pivot table berhasil dibuat dan disimpan!")
```
